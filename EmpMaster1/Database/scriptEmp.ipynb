# [dbo].[InsertUpdateEmployee]

```sql
/****** Object:  StoredProcedure [dbo].[InsertUpdateEmployee]    Script Date: 11-08-2022 18:00:25 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE Procedure [dbo].[InsertUpdateEmployee]    
(    
@Id integer,    
@Name nvarchar(50),    
@Email nvarchar(50),    
@State nvarchar(50),    
@City nvarchar(50),
@Phone_No nvarchar(50),
@Address nvarchar(50),
@Agree bit=Null,
@Action varchar(10)    
)    
As    
Begin    
if @Action='Insert'    
Begin    
 Insert into Employee(Name,Email,[State],City,Phone_No,Address,Agree) values(@Name,@Email,@State,@City,@Phone_No,@Address,@Agree);    
End    
if @Action='Update'    
Begin    
 Update Employee set Name=@Name,Email=@Email,[State]=@State,City=@City,Phone_No=@Phone_No,Address=@Address,Agree=@Agree where EmployeeID=@Id;    
End      
End  
  
GO
```
